```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('info-rrhh-por-genero1.csv')
df = df.pivot(index='Jerarquia', columns='Empleo', values='Varones')
df.plot(kind='bar', figsize=(15, 8), color=['red', 'blue', '#e37827', '#275444'])

plt.xlabel('Jerarquia')
plt.ylabel('Varones')
plt.title('Empleos Jerarquicos Varones en la Ciudad de Bs. As.')
plt.show()
```
